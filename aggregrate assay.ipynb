```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
path = "C:/Users/ZENGW6/Desktop/tox21/tox21"

text_files = glob.glob(path + "/**/*[aggregrated].txt", recursive = True)

# print(text_files)

# pd.DataFrame(text_files)
```

```python
df = []
for i in range(len(text_files)):
    df2 = pd.read_csv(text_files[i], delimiter = "\t", index_col=False)
    df.append(df2)
```

```python
file = df[0].loc[df[0]['SAMPLE_DATA_TYPE'] == 'viability']
file
```

```python
print("Assay Outcome\n",file.ASSAY_OUTCOME.value_counts())
print('\nChannel Outcome\n', file.CHANNEL_OUTCOME.value_counts())
print('\nPurity Rating\n', file.PURITY_RATING.value_counts())
print('\nPurity Rating 4M\n', file.PURITY_RATING_4M.value_counts())
```

```python
# Test Plot AC50
import matplotlib.pyplot as plt
molwt = file[['ASSAY_OUTCOME', 'AC50']]
fig, ax = plt.subplots(figsize=(10,6))

for label, molwt in molwt.groupby('ASSAY_OUTCOME'):
    molwt.AC50.plot(kind="kde", label=label)
plt.legend()
plt.title('Kernel Density Plot for AC50')
plt.show()
```

```python
molwt = file[['ASSAY_OUTCOME', 'EFFICACY', 'CURVE_RANK', "AC50"]]
fig, ax = plt.subplots(1, 2, figsize=(10,6))

ax = ax.ravel()

for label, molwt in molwt.groupby('ASSAY_OUTCOME'):
#     molwt.EFFICACY.plot(kind="kde", label=label)
    molwt.EFFICACY.plot(ax = ax[0], kind="kde", label=label)
    molwt.CURVE_RANK.plot(ax = ax[1], kind = 'hist', label = label, bins = 30)

    ax[0].set_title('Kernel Density Plot for EFFICACY')
    ax[0].legend()
    ax[1].set_title('Histogram for CURVE_RANK')
    ax[1].legend()
plt.show()
```
