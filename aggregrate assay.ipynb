```python
# Test Plot AC50 assay outcome
import matplotlib.pyplot as plt
molwt = file[['ASSAY_OUTCOME', 'AC50']]
fig, ax = plt.subplots(figsize=(10,6))
try:
    for label, molwt in molwt.groupby('ASSAY_OUTCOME'):
        molwt.AC50.plot(kind="kde", label=label)
    plt.legend()
    plt.title('Kernel Density Plot for AC50')
    plt.show()
except:
    pass
```

```python
molwt = file[['ASSAY_OUTCOME', 'EFFICACY', 'CURVE_RANK', "AC50"]]
fig, ax = plt.subplots(1, 2, figsize=(10,6))

ax = ax.ravel()

for label, molwt in molwt.groupby('ASSAY_OUTCOME'):
#     molwt.EFFICACY.plot(kind="kde", label=label)
    molwt.EFFICACY.plot(ax = ax[0], kind="kde", label=label)
    molwt.CURVE_RANK.plot(ax = ax[1], kind = 'hist', label = label, bins = 30)

    ax[0].set_title('Kernel Density Plot for EFFICACY')
    ax[0].legend()
    ax[1].set_title('Histogram for CURVE_RANK')
    ax[1].legend()
plt.show()
```
